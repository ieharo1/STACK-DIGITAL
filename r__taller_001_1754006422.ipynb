```python
import pandas as pd
import matplotlib.pyplot as plt
```

Paso 3: Cargamos datos y trabajaremos con los datos del Titanic

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Explorar y limpiar datos
# Ver informacion general
df.info()
```

# Eliminar columnas innecesarias

```python
df.drop(['Cabin', 'Ticket'], axis=1, inplace=True)
```

```python
# Llenar valores nulos en Edad con la media
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
# Llenar valores nulos en Edad con la media
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
# Verificar cambios al preguntar si hay valores nulos
df.isnull().sum()
```

```python
#Paso 5: Visualizar los datos
# Histograma de edades
plt.hist(df['Age'], bins=20, color='skyblue')
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de barras por clase
df['Pclass'].value_counts().plot(kind='bar', color='orange')
plt.title('Cantidad de Pasajeros por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```
